## simulate coin flips

```python
import numpy as np

# Generate the sum of k coin flips, repeat that n times
def generate_counts(k=1000, n=100):
    # generate a (k, n) matrix of value
    x = 2 * (np.random.rand(k, n) > 0.5) - 1
    return np.sum(x, axis=0)
```

```python
import matplotlib.pyplot as plt

test = generate_counts(1000, 1000)
plt.hist(test)
plt.grid()
plt.xlim(-1000, 1000)
plt.xlabel("sum")
plt.ylabel("count")
plt.title("Histogram of coin flip sum when flipping a fair coin 1000 times")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(3,1,1)
ax.set_title("100 flips, bound = +- 40.0")
ax.hist(generate_counts(100, 1000))
ax.axvline(x=-40, color='r')
ax.axvline(x=40, color='r')
plt.grid()

ax = fig.add_subplot(3,1,2)
ax.set_title("1000 flips, bound = +- 126.5")
ax.hist(generate_counts(1000, 1000))
ax.axvline(x=-126.5, color='r')
ax.axvline(x=126.5, color='r')
plt.grid()

ax = fig.add_subplot(3,1,3)
ax.set_title("10000 flips, bound = +- 400.0")
ax.hist(generate_counts(10000, 1000))
ax.axvline(x=-400.0, color='r')
ax.axvline(x=400.0, color='r')
plt.grid()

fig.tight_layout()
plt.show()
```
